```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fn(x):
    return 3*x**2-4*x+5

def derfn(x):
   return 6*x-4
```

```python
x = np.linspace(-2.5,4,100)
y = fn(x)
plt.plot(x,y)
plt.show()
```

```python
xm = -2
alpha = 0.1
xml = [xm]
```

```python
for i in range(20):
    # here 20 is no of terms which is generally known as epochs
    xm = xm-alpha*derfn(xm)
    xml.append(xm)
```

```python
plt.plot(x,y)
plt.plot(xml,fn(np.array(xml)))
```
