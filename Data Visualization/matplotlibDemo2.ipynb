```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(-2,2,1000)
ysin = np.sin(2*np.pi*x)
ycos = np.cos(2*np.pi*x)

ytan = np.tan(2*np.pi*x)
ycot = ycos/ysin
fig,axes = plt.subplots(2,2,figsize=(12,6))
# fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2,figsize=(12,6))
# figsize = (weight,height)

axes[0,0].plot(x,ysin)
axes[0,0].grid()
axes[0,0].set_title("Sin Wave")

axes[0,1].plot(x,ycos)
axes[0,1].grid()
axes[0,1].set_title("Cos Wave")

axes[1,0].plot(x,ytan)
axes[1,0].grid()
axes[1,0].set_ylim((-5,5))
axes[1,0].set_title("Tan Wave")

axes[1,1].plot(x,ycot)
axes[1,1].grid()
axes[1,0].set_ylim((-5,5))
axes[1,1].set_title("Cot Wave")

plt.suptitle("Tignometric functions")
plt.subplots_adjust(hspace=0.3)
plt.show()
```

```python
x = np.array([10,20,30])
y2024 = np.array([80,60,100])
y2025 = np.array([90,80,90])

# plt.bar(x+1.25,y2024,width=2.5,tick_label=['Gold','Silver','Platinum'],label='2024')
# plt.bar(x-1.25,y2025,width=2.5,tick_label=['Gold','Silver','Platinum'],label='2025')

plt.bar(x+1.25,y2024,width=2.5,label='2024')
plt.bar(x-1.25,y2025,width=2.5,label='2025')

plt.xticks(x,['Silver','Gold','Platinum'])
plt.legend()
plt.title('Metal Prices')
plt.ylabel('Prices')
plt.show()
```

```python
x = np.array([10,20,30])
y2024 = np.array([80,60,100])
y2025 = np.array([75,80,90])

plt.barh(x-1.25,y2024,height=2.5,label='2024')
plt.barh(x+1.25,y2025,height=2.5,label='2025')

plt.yticks(x,['Gold','Silver','Platinum'])
plt.legend()
plt.show()
```
